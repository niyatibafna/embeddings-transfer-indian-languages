```python
with open("tag_distributions/tag_dist_l.json", "r") as l_file:
    tag_dist_l = json.load(l_file)
    
with open("tag_distributions/tag_dist_h.json", "r") as h_file:
    tag_dist_h = json.load(h_file)
```

```python
def find_cost(morph_l, morph_h):
    ned = editdistance.eval(morph_l, morph_h)/max(len(morph_l), len(morph_l))
    return ned if ned < 0.5 else 10
        
```

```python
def populate_cost_array(morphlist_l, morphlist_h):
    A = np.zeros([len(morphlist_l), len(morphlist_h)])
    for i_l, m_l in enumerate(morphlist_l):
        if i_l % 1000 == 0:
            print(i_l)
        for i_h, m_h in enumerate(morphlist_h):
            A[i_l][i_h] = find_cost(m_l, m_h)
    return A
```

```python
cost_array = populate_cost_array(list(tag_dist_l.keys()), list(tag_dist_h.keys()))
```

```python
row_ind, col_ind = linear_sum_assignment(cost_array)
```

```python
a = list(tag_dist_l.keys())
b = list(tag_dist_h.keys())
```

```python
for l, hidx in enumerate(col_ind[:100]):
    print(a[l])
    print(b[hidx])
    print("\n\n\n")
#     cost_array
```

```python
morph_mapping = dict()
for ridx in range(len(cost_array)):
    morph_mapping[ridx] = int(np.argmin(cost_array[ridx]))
```
