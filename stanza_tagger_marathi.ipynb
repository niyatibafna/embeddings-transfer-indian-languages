```python
morph_file = open("marathi_data/mr/mr.tok.txt").read().split("\n")[1000:1050]
gold_tag = open("marathi_data/mr/mr.tok.gold_tagged.txt", "w")
# data = open("marathi_data/mr/mr.segmented.EOLmarked.txt", "r").read().split("E O L")[700:750]
# data = [re.split(' |\n', ss)[1:-1] for ss in data]

for line in morph_file:
#     word = line
    print(line)
    doc = nlp(line)
    text = [token.text for sent in doc.sentences for token in sent.tokens]
    tags = []
    for sent in doc.sentences:
        for token in sent.tokens:
            if len(token.id)==1:
                tags+=(word.upos for word in token.words)
            else:
                tags.append("MISSING")
    assert(len(text) == len(tags))
    print(*list(zip(text, tags)), sep = "\n")
    for word, tag in list(zip(text, tags)):
        gold_tag.write(word+"\t"+tag+"\n")
    gold_tag.write("E O L\n")
    
    
#     print([token.parent.text for sent in doc.sentences for token in sent.words])

#     print(doc.sentences[0].words[0].text, "\t", doc.sentences[0].words[0].upos)
#     print(*[f'word: {word.text+" "}\tupos: {word.upos}\txpos: {word.xpos}' for sent in doc.sentences for token in sent.tokens for word in token.words], sep='\n')
    print("\n")
gold_tag.close()
```

```python
import re
data = open("marathi_data/mr/mr.segmented.EOLmarked.txt", "r").read().split("E O L")[700:750]
data = [re.split(' |\n', ss)[1:-1] for ss in data]
```

```python
tagged_data = open("marathi_data/mr/mr.segmented.gold_tagged.txt", "w")
for sent in data:
    tagged_data.write(" ".join(sent)+"\n")
    for word in sent:
        tagged_data.write(word+"\n")
    tagged_data.write("E O L\n")
tagged_data.close()
```
