```python
import stanfordnlp
```

```python
nlp = stanfordnlp.Pipeline(lang="hi", processors='tokenize,mwt,pos')
```

```python
tok_file = open("hindi_data/hin.tok.subset100.txt", "r")
tag_file = open("hindi_data/hin.tag.subset100.txt", "w")
seg_file = open("hindi_data/hin.tobesegmented.subset100.txt","w")
for line in tok_file:
    line = line.strip("\n")
    doc = nlp(line)
    for sent in doc.sentences:
        for word in sent.words:
            tag_file.write(word.text+"\t"+word.upos+"\n")
            seg_file.write(word.text+"\n")
tag_file.close()
seg_file.close()
tok_file.close()
```
